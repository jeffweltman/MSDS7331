```python
%config IPCompleter.greedy=True
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
# all_hs_2017 = pd.read_csv("https://raw.githubusercontent.com/jakemdrew/EducationDataNC/master/2017/School%20Datasets/PublicHighSchools2017.csv")
ml_hs_2017_all = pd.read_csv("https://github.com/jakemdrew/EducationDataNC/raw/master/2017/Machine%20Learning%20Datasets/PublicHighSchools2017_ML.csv")
#ml_hs_2017['sat_avg_level'] = np.where(ml_hs_2017['sat_avg_score_num'] >=1 & ml_hs_2017['sat_avg_score_num'] <=900,'Low',np.where(ml_hs_2017['sat_avg_score_num'] >900 & ml_hs_2017['sat_avg_score_num'] <= 1200,'Medium','High'))

ml_hs_2017 = ml_hs_2017_all.loc[ml_hs_2017_all['sat_avg_score_num'] >0]
```

```python
pd.options.mode.chained_assignment = None

ml_hs_2017.loc[(ml_hs_2017['sat_avg_score_num'] >= 1) & (ml_hs_2017['sat_avg_score_num'] < 900),'sat_avg_level'] = 'Low'
ml_hs_2017.loc[(ml_hs_2017['sat_avg_score_num'] >= 900) & (ml_hs_2017['sat_avg_score_num'] < 1200),'sat_avg_level'] = 'Medium'
ml_hs_2017.loc[ml_hs_2017['sat_avg_score_num'] >= 1200,'sat_avg_level'] = 'High'

# Get sizes
print("Rows: " + str(ml_hs_2017.shape[0]) + " ~~~~ Columns: " + str(ml_hs_2017.shape[1]))

# output
ml_hs_2017.to_csv(path_or_buf='./ml_hs_2017_plus.csv', sep=',')
```
