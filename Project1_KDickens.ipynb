# Data Meaning Type
The North Carolina Educational dataset pertaining to high schools consists of 403 attributes describing a broad range of data points from school overview, attendance and enrollment, student performance, safety and environment, and access to technology.  This section will describe the meaning and type of data included in this project.  For each attribute in the dataset this section will define the attribute and state the units for a measurement of scale and the type of values used.

## Educator Profile
The primary purpose of teh educator profile data is to examine the the fitness of the educators responsible for the students learning activites.  This covers the teachers educationa and professional licensing, their relevant work experience, as well as whether they meet or exceed certain state standards.

### Education & Licensing
Education and licensing data looks at the percentage of teachers professionally licensed as well as the advanced education levels of teachers.  Licensing comes in two forms with the standard teacher's license and then the National Board Certification which is a more prestigious accomplishment.  Education focuses solely on the percent of teachers with advanced degrees (masters degree or higher).  Six continuous attributes cover the important facets of educator education and licensing.

## Performance Profile
The performance profile contains data that indicates how the school performs at academic standards such as Math, Science, and Reading.  This includes additional courses taken for advanced placement and career focused classes.  The key indicators for school performance is teh School Performance Grade and teh Education Value added System.  Also included are common test results for high school students such as ACT and SAT.

### School Performance Grade (SPG)
School Performance Grade, or SPG, is an aggregate score that indicates how well a school performs academically.  It takes into consideration end of grade, end of course, graduation rates, and college and workplace readiness measures.  It also factors in school growth measures such as EVAAS.  SPG Score contains thirteen attributes describing the key performance metric for North Carolina schools.  Of these thirteen only the SPG Score itself is continuous.  Other variables are one-hot encoded variables to describe scoring outcomes.

This section will analyze the simple statistics for several performance indicators across the spectrum of testing, graduation, and academic success.

```python
# import ipython to allow display of multiple outputs in Jupyter cells.
from IPython.display import display 

# read in file
df=pd.read_csv('PublicHighSchools2017_ML.csv')

# create data frame object with desired variables & rename columns
df_per=df.loc[:, ['SPG Score', 'NC Math 1 Score', 'English II Score', 'Biology Score', 'ACT Score', 'ACT WorkKeys Score', 'ACTMath_ACTBenchmark_All', 'ACTScience_ACTBenchmark_All', 'ACTWorkKeys_SilverPlus_All', 'ACTWriting_ACTBenchmark_All', 'sat_avg_score_num', 'lea_sat_avg_score_num', 'sat_participation_pct', 'lea_sat_participation_pct', '4-Year Cohort Graduation Rate Score']]
df_per.columns=['SPG', 'Math1', 'English2', 'Biology', 'ACT', 'ACTWorkkeys', "ACTMath", 'ACTScience', 'ACTWorkkeys_Silver', 'ACTWriting', 'SAT', 'LEA_SAT', 'SAT_Part', 'LEA_SAT_Part','4Year_Grad']

# Subset SPG grades
df_per_c=df.loc[:,['SPG Grade_A+NG', 'SPG Grade_B', 'SPG Grade_C', 'SPG Grade_D']]

# Print summary statistics and sum of SPG by grade
print('Summary Statistics')
display(df_per.describe())
print('Schools by SPG Grades')
display(df_per_c.sum())
```

The evaluation of test scores shows several interesting trends.  North Carlolina students score significantly lower on science than the other subjects with the percent of students scoring a 17 or better a full 6 points lower than math or writing.  Likewise math and science courses reflect lower scores than english across the state.  Surprisingly, given the percentage scale of 0-100 all of the test scores show relatively high variance with standard deviations of 10-15 incredibly common.

The best overall indicator of school performance in North Carolina are the School Performance Grades which take into account standardized tests, academic success, graduation rates, and other factors.  A simple summary shows that out of 470 schools roughly 26 scored As, while 174 scored Bs.  That means roughly 43% of all publich high schools in North Carolina performed very well.

```python
# Set properties for histogram
sb.set(context='notebook', style='whitegrid', font_scale=1.5)
fig, ax=plt.subplots(figsize=(15,8))

# Plot histogram of ACT Subject score percentages
display(sb.distplot(df_per['ACTMath']))
display(sb.distplot(df_per['ACTScience']))
display(sb.distplot(df_per['ACTWriting']))

# Set plot title, labels and legend
plt.xlim(0, 100)
plt.xlabel('ACT Score as percent of Participants Receiving a 17 or Higher')
plt.ylabel('Frequency')
plt.title('ACT Score Distribution by Subject')
plt.legend(['Math', 'Science', 'Writing'])
```

The histogram of ACT subject results clearly shows that that students perform slightly better at writing as opposed to math and science.  Note that the scale is constrained between 0 and 100 because these are percentages.

## Environment
The environment section will look at exposure to crime and the availability to educational resources.  The attributes chosen here are: Crimes committed per 100 students at the school, LEA, and State levels, the number of books per student, and whether the school is a Bring Your Own Device school.

```python
# import ipython to allow display of multiple outputs in Jupyter cells.
from IPython.display import display 

# read in file
df=pd.read_csv('PublicHighSchools2017_ML.csv')

# create data frame object with desired variables & rename columns
df_env=df.loc[:, ['crime_per_c_num', 'lea_crime_per_c_num', 'st_crime_per_c_num', 'books_per_student', 'lea_books_per_student', 'Byod_Yes']]
df_env.columns=['Crime', 'LEA_Crime', 'ST_Crime', 'Books', 'LEA_Books', 'BYOD']

#Print summary statistics and a sum of the BYOD schools = 1
print('Summary Statistics')
display(df_env.describe())
print('Number of BYOD Schools')
display(df_env['BYOD'].sum())
    
```

The crime statistics show there is much greater variance between individual schools than there is between LEAs.  Interestingly enough the high schools show an overall lower mean and median value for crime per 100 students than either the LEA or the state.  One would expect that the high schools would be higher, or at least closer to the LEA and/or state values.  

Students at high school in North Carolina seem to have lower mean and median number of books per student than the overall LEAs.  They also show a really high standard deviation showing significant variances between individual schools. This could be due to internet resources and older children's increased ability to access such resources while younger kids benefit more from access to physical book media. The range when compared to the interquartile range and mean or median, indicate the presence of significant outliers. This bears further investigation if included in any future model.

Lastly we can see that, of the 470 public high schools in the state of North Carolina, less than half are Bring Your Own Device schools. This means that, in over half the schools students are issued an internet capabale device such as a laptop or tablet.

Let's take a look at the same data, but visualized more graphically.

```python
#Set properties for histogram
sb.set(context='notebook', style='whitegrid', font_scale=1.5)
fig, ax=plt.subplots(figsize=(15,8))

#plot out histograms for Crime and LEA Crime
display(sb.distplot(df_env['Crime']))
display(sb.distplot(df_env['LEA_Crime']))

# Set plot title, labels and legend
plt.xlim(0, None)
plt.xlabel('Crime per 100 Students')
plt.ylabel('Frequency')
plt.title('Crime per  100 Students in Public High Schools versus Districts')
plt.legend(['Crime', 'LEA Crime'])
```

Here we can see the crime statistics as a histogram where the blue represents individual schools and the green LEAs.  As expected from the tabular statistics the individual schools show a long-tailed right skewed distribution while the LEA histogram shows a more normal distribution with a much slighter skew.
